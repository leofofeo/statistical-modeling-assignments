##### b. Next, fit a regression model with both explanatory variables. Has the addition of SquareFeet changed our estimate of the Distance relationship to Adj2007? Use both the estimated coefficient and $R^2$ in your answer.

```r
model <- lm(adj2007 ~ distance + squarefeet, data = rails_trails)
summary(model)
```

Yes, the addition of SquareFeet has changed the distance estimate. Additionally, $R^2$ has increased to .7655 from .2374, which is a considerable increase, and suggests that square feet is a stronger and more relevant predictor of house prices than distance.

# Problem 3.19

##### a. Fit a multiple regression model with Animus depending on the four predictors. Report the fitted prediction equation


```r
animus_data <- read.csv("./data/RacialAnimus.csv")
model <- lm(Animus ~ Age65Plus + BachPlus + Black + Hispanic, data = animus_data)
summary(model)
```

Based on the model summary, our prediction equation is
$$Animus = 124.3195 + 1.5189(\text{Age65Plus}) - 2.0539(\text{BachPlus}) + 0.5640(\text{Black}) - 0.5779(\text{Hispanic})$$

##### b. Prepare appropriate residual plots and comment on the conditions for inference

```r
plot(model$fitted.values, model$residuals,
     xlab="Fitted Values", ylab="Residuals",
     main="Residuals vs Fitted")
abline(h=0, col="red", lty=2)
```

For the most part, we do see some random and even distribution around the horizontal line, although there are definitely some outliers toward the right hand of our plot.

```r
# Normal Q-Q plot
qqnorm(model$residuals)
qqline(model$residuals, col="red")
```

Similarly, we see here that most points follow the diagonal line, indicating that the bulk of our residuals are following a normal distribution. The deviation we see toward the right mirrors the pattern we saw in our fitted vs residuals plot, indicating that there are some outliers where scores were much higher than our model predicted.
